```python
import pandas as pd #panda is Use for reading data from csv files, storing data in dataframes, helps in handaling data in rows and columns plotting graphs.
import seaborn  as sns # seaborn is used for ploting graphs in python. 
import matplotlib.pyplot as plt #this is needed to create graphs and charts from data.
```

```python
# Load the dataset
def load_dataset(file_path): #function to load dataset from csv file.
    return pd.read_csv(file_path)  #reading csv file using pandas and returning dataframe.
df=load_dataset("Health_heart_experimental.csv") #loading dataset from csv file named "Health_heart_experimental.csv"
df.head(11)
```

# Line chart

```python
df_group = df[df['age'].between(40, 60)]

sns.lineplot(x='age', y='SysBP', data=df_group, color='blue')
# sns.lineplot(x='age', y='DiaBP', data=df_group, color='orange')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.title('Systolic BP Trend (Ages 40-60)')
plt.show()
```

The given line chart shows the Systolic Blood Pressure of 40 to 60 age group people. This chart shows how the pressure in arteries when heart beats changes as you grow older. We can see slightly increase in the SYSBP as the age increases. For people of age group 40 their SYSBP is around 125.50. But as the age increase we can see rise in SYSBP. SYSBBP for people of age group is 125.80 which is increse by 0.239%.  

```python
df_group = df[df['age'].between(50, 69)]

# sns.lineplot(x='age', y='SysBP', data=df_group, color='blue')
sns.lineplot(x='age', y='DiaBP', data=df_group, color='orange')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.title('Diastolic BP Trend (Ages 50-69)')
plt.show()
```

Just like in Systolic Blood Pressure we can see similar trend for 50 to 69 age group. The diastolic Blood Pressure Increases as the age increases. The highest Diastolic Blood Pressure is 96.3 which is of age 68 group while the DiaBP of age group is 95.7. Which is increase by 0.63%

```python

df_group = df[df['age'].between(20, 40)]
pulse_pressure = df_group['SysBP'] - df_group['DiaBP']
sns.lineplot(x='age', y=pulse_pressure, data=df_group, color='green')
plt.xlabel('Age')
plt.ylabel('Pulse Pressure')
plt.title('Pulse Pressure Trend (Ages 20-40)')
plt.show()
```

The given Line Chart shows the trend of Pulse Pressure of age group of 20 to 40. Here we can see peak Pulse Pressure(PP) is of age group 29 to 30, as this is the prime age for any human beings. The PP for age group 20 is 29.5 and for age group 40 is 30.2 which is not much increase. But we can see for prime age group the PP is 30.4 which great increased from 29.5.

# Bar Plot

```python
sns.countplot(data=df, x='sex', hue='indication', palette='pastel')
plt.title('Distribution of Indication by Sex')
plt.xlabel('sex(0 = female , 1 = male )')
plt.ylabel('Count of individuals')
plt.show()
```

In medical terminology, "indication" typically refers to a diagnostic marker suggesting the presence of disease. In this dataset context, incdication is likely a binary classification 0 indicating no heart disease or Healthy while 1 indicating heart disease present or at risk. In the given bar graph x axis is gender and y axis is the number of people. As per female ~18,800 had no indication (0), while ~17,100 did (1). And for male ~18,900 had no indication (0), while ~16,800 did (1). The number of male who may are at risk is less than female by 1.51%.

# Pie Chart

```python
def categorize_bmi(BMI):
    if BMI < 18.5:
        return 'Underweight'
    elif 18.5 <= BMI < 24.9:
        return 'Normal weight'
    elif 25 <= BMI < 29.9:
        return 'Overweight'
    else:
        return 'Obesity'
    
df['BMI_Category'] = df['BMI'].apply(categorize_bmi)
BMI_counts = df['BMI_Category'].value_counts()
plt.pie(BMI_counts, labels=BMI_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Distribution of BMI Categories')
plt.show()
```

This pie chart illustrates the distribution of BMI categories in the heart health dataset, revealing a population with significant health concerns. Obesity is the dominant category at nearly 50% , representing almost half of all individuals in the study. The overweight category accounts for 21.8%, meaning that combined, over 71% of the population has a BMI above the normal range. Normal weight individuals comprise only 20.3% of the dataset, while underweight represents the smallest group at just 8.3%. This distribution suggests a high-risk population for cardiovascular disease, as excess weight is a major risk factor for heart conditions, hypertension, and metabolic disorders. The prevalence of obesity and overweight status in this dataset likely correlates with elevated values in other health metrics like blood pressure and may be a strong predictor of positive indication values (presence of heart disease).

# Box Plot

```python
sns.boxplot(x='indication', y='HR', data=df, color='skyblue')
plt.title('Heart Rate Distribution by Indication')
plt.xlabel('Indication (0 = No, 1 = Yes)')
plt.ylabel('Heart Rate (HR)')
plt.show()
```

This box graph illustrates the distribution of heart rate measurements in those without heart disease (indication = 0) compared to those with heart disease (indication = 1), which shows surprisingly similar figures in regard to cardiovascular measures in both cases. The medians of their heart rate measurements are almost the same in both cases, at around 138 to 140 beats per minute, with their boxes ranging from 110 to 170 measurements of beats per minute. The whiskers of the boxes are also similar in range, which spanned from around 80 beats per minute at their lowest points to 200 measurements at their highest points, showing similar dispersions of heart rate measurements in both cases.

The highest potential in this graph occurs at the higher ends of the BMI (around 33 to 35) which nearly everybody is obese. Also, its observable that theres a significant amount of data distributed at the lower ends of the BMI (around 15 to 20).

# Scatterplot

```python
df_40 = df[df['age'] == 40]
df_sample = df_40.sample(n=100, random_state=42)
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_sample, 
    x='BMI', 
    y='HR', 
    hue='indication', 
    style='indication', 
    s=120,              
    palette='Set1'      
)
plt.title('BMI vs Heart Rate at Age 40')
plt.xlabel('BMI')
plt.ylabel('Heart Rate (HR)')
plt.show()
```

This scatter plot shows BMI versus heart rate for 40-year-olds, with red dots for healthy people and blue X's for those with heart disease. The pattern is pretty clear: healthy individuals cluster in the lower BMI range (17-30), while those with heart disease dominate the higher end (30-40). There's a visible separation around BMI 30, where blue markers become much more common than red ones.
